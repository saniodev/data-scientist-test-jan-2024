```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def carregar_dados():
    area_agosto_mg_por_municipio = pd.read_csv('dados/area_agosto_mg_por_municipio.csv', index_col=0)
    area_setembro_mg_por_municipio = pd.read_csv('dados/area_setembro_mg_por_municipio.csv', index_col=0)
    area_por_bioma_agosto = pd.read_csv('dados/area_por_bioma_agosto.csv', index_col=0)
    area_por_bioma_setembro = pd.read_csv('dados/area_por_bioma_setembro.csv', index_col=0)
    areas_totais = pd.read_csv('dados/areas_totais.csv')
    correlacao = pd.read_csv('dados/correlacao.csv', index_col=0)

    return area_agosto_mg_por_municipio, area_setembro_mg_por_municipio, area_por_bioma_agosto, area_por_bioma_setembro, areas_totais, correlacao

def plotar_graficos(area_agosto_mg_por_municipio, area_setembro_mg_por_municipio, area_por_bioma_agosto, area_por_bioma_setembro, areas_totais, correlacao):
    sns.set(style="whitegrid")

    plt.figure(figsize=(12, 6))
    sns.barplot(x='area_km2', y='nome_municipio', data=area_agosto_mg_por_municipio)
    plt.title('Desmatamento por Município em Agosto de 2022')
    plt.xlabel('Área Desmatada (km²)')
    plt.ylabel('Município')
    plt.show()

    plt.figure(figsize=(10, 8))
    sns.scatterplot(x='area_km2', y='nome_municipio', size='area_km2', sizes=(50, 500), data=area_setembro_mg_por_municipio)
    plt.title('Desmatamento por Bioma em Setembro de 2022')
    plt.xlabel('PIB Nominal')
    plt.ylabel('População')
    plt.show()

    plt.figure(figsize=(8, 6))
    sns.heatmap(correlacao, annot=True, cmap='coolwarm')
    plt.title('Correlação entre Variáveis')
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Bioma', y='area_km2', data=area_por_bioma_agosto)
    plt.title('Distribuição do PIB Nominal por Bioma em Agosto de 2022')
    plt.xlabel('Bioma')
    plt.ylabel('PIB Nominal')
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.barplot(x='Bioma', y='area_km2', data=area_por_bioma_setembro)
    plt.title('Contagem de Registros por Bioma em Setembro de 2022')
    plt.xlabel('Bioma')
    plt.ylabel('Contagem')
    plt.show()

if __name__ == "__main__":
    area_agosto_mg_por_municipio, area_setembro_mg_por_municipio, area_por_bioma_agosto, area_por_bioma_setembro, areas_totais, correlacao = carregar_dados()

    plotar_graficos(area_agosto_mg_por_municipio, area_setembro_mg_por_municipio, area_por_bioma_agosto, area_por_bioma_setembro, areas_totais, correlacao)
```
